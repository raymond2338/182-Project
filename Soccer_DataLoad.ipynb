```python
for index in tqdm(range(len(x_y))):
    for player in player_apis:
        temp = player_attributes[(player_attributes["player_api_id"]==x_y.loc[index, player]) & \
                                 (player_attributes['date'] < x_y.loc[index, "date"])]
        temp_stats = temp.iloc[temp['date'].values.argmax(), 4:]
        for stat_col in temp_stats.index:
            x_y.loc[index, player + "_" + stat_col] = temp_stats[stat_col]
```
